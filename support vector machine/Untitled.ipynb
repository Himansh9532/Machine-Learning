```python
x = df.drop("target_class" , axis = 1)
y = df["target_class"]
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y , test_size = 0.20 , random_state = 42)
```

# 11. Feature Scalling

```python
cols = x_train.columns
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train = pd.DataFrame(x_train , columns = [cols])
```

```python
x_test = pd.DataFrame(x_test , columns = [cols])
```

```python
x_train.describe()
```

```python
# 12. Run SVM with default hyperparameters
# Default hyperparameter means C=1.0, kernel=rbf and gamma=auto among other parameters.
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.metrics import accuracy_score
```

```python
svc = SVC()
```

```python
svc.fit(x_train , y_train)
```

```python
y_pred = svc.predict(x_test)
```

```python
print("Model accuracy score with default hyperparameters : {0:0.4f}".format(accuracy_score(y_test , y_pred)))
```

# 16. Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
print("confusion mattrix" , cm)
print("True Positive" , cm[0,0])
print("True Negative" , cm[1,1])
print("False Positive" , cm[0,1])
print("False Negative" , cm[1,0])
```

```python
#Visualize  confusion matrix with seaborn heatmap
cm_matrix = pd.DataFrame(data = cm , columns = ["Actual Positive1" , "Actual Negative : 0"], index = ["Predict Positive:1" , "Predict Negative:0"])
```

```python
sns.heatmap(cm_matrix , annot = True , fmt = "d" , cmap = "YlGnBu")
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test , y_pred))
```

```python
TP = cm[0,0]
TN = cm[1,1]
FP = cm[0,1]
FN = cm[1,0]
```

```python
classification_accuracy = (TP+TN)/float(TP+TN+FP+FN)
```

```python
print(classification_accuracy)
```

```python
classification_error = (FP+FN)/float(TP+TN+FP+FN)
print(classification_error)
```

# Precision

```python
precision = TP / float(TP+FP)
print("Precision" , precision)
```

# Recall

```python
recall = TP/float(TP+FN)
print("Recall or sensitivity" , recall)
```

```python
true_positive_rate = TP / float(TP + FN)
print('True Positive Rate : {0:0.4f}'.format(true_positive_rate))
```

```python
false_positive_rate = FP / float(FP + TN)
print('False Positive Rate : {0:0.4f}'.format(false_positive_rate))
```

```python
specificity = TN / (TN + FP)
print('Specificity : {0:0.4f}'.format(specificity))
```

# ROC CURVE (RECEIVER OPERATING CHARACTERSTICS CURVE)

```python
from sklearn.metrics import roc_curve
```

```python
fpr,tpr,thresholds = roc_curve(y_test , y_pred)
plt.figure(figsize = (6,4))
plt.plot(fpr,tpr, linewidth = 2)
plt.plot([0,1] , [0,1],"k--")
plt.title('ROC curve for Predicting a Pulsar Star classifier')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.show()
```
